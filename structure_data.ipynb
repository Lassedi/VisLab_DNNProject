```python
import shutil

# find and remove classes with too little images per class
_,d,_ = next(iter(os.walk("./Data/img_align_celeba/class_data/")))
count = 0
imgC_cutoff = 2
rm = False

for dir in range(len(d)):
    # print("class: {}, index: {}".format(d[dir], dir))
    r,_,f = next(iter(os.walk("./Data/img_align_celeba/class_data/{}".format(d[dir]))))
    if len(f)<imgC_cutoff:
        count += 1
        print("Class: {} contains only {} images".format(r, len(f)))
        if rm: # remove directory if less than cutoff
            shutil.rmtree(r)
    

print("Total Classes with less than {} images per class: {}".format(imgC_cutoff, count)) 
```
